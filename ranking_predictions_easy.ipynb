## AR(10) Market Cap Prediction — Target: April 30, 2025

Predict each stock's market cap 30 trading days ahead using an AR(10) model trained exclusively on data available before April 1, 2025. Predicted market caps are ranked to evaluate how well rank order is preserved.

| Parameter | Value |
|---|---|
| Training window | 2024-01-01 → 2025-03-31 (~315 trading days) |
| Forecast horizon | 30 trading days |
| Target date | ~April 30, 2025 |

### Step 1 — Pivot & Split

Set `Dates` as the index so each column is a stock ticker, then split into train (≤ March 31, 2025) and test (> March 31, 2025).

```python
import numpy as np
from scipy.stats import spearmanr

# Dates as index; each column is a stock ticker
mktcap_df = df.set_index("Dates")
mktcap_df.index = pd.to_datetime(mktcap_df.index)
mktcap_df = mktcap_df.sort_index()

cutoff = pd.Timestamp("2025-03-31")
train = mktcap_df[mktcap_df.index <= cutoff]   # ~315 trading days
test  = mktcap_df[mktcap_df.index > cutoff]    # remaining days

print(f"Train rows: {len(train)}, Test rows: {len(test)}, Stocks: {train.shape[1]}")
```

### Step 2 — AR(10) Fit & 30-Step Forecast

For each stock with at least 12 non-NaN observations in the training window, fit an AR(10) model on **first-differences** via OLS, then iteratively forecast 30 steps ahead. The predicted market cap equals the last observed market cap plus the cumulative sum of the forecasted differences. Stocks with insufficient history are skipped.

```python
LAGS  = 10
STEPS = 30

def fit_forecast_ar(series, lags=LAGS, steps=STEPS):
    s = series.dropna().values
    diffs = np.diff(s)
    if len(diffs) <= lags:
        return None
    X = np.column_stack([diffs[i:len(diffs)-lags+i] for i in range(lags)])
    y = diffs[lags:]
    X_b = np.c_[np.ones(len(y)), X]
    coeffs, _, _, _ = np.linalg.lstsq(X_b, y, rcond=None)
    history = list(diffs[-lags:])
    for _ in range(steps):
        x = np.r_[1.0, history[-lags:]]
        history.append(float(coeffs @ x))
    cumulative_delta = sum(history[lags:])
    last_mktcap = series.dropna().iloc[-1]
    return last_mktcap + cumulative_delta

predictions = {}
for stock in train.columns:
    s = train[stock]
    if s.dropna().shape[0] < LAGS + 2:
        continue
    pred = fit_forecast_ar(s)
    if pred is not None:
        predictions[stock] = pred

print(f"Stocks with valid forecasts: {len(predictions)}")
```

### Step 3 — Actual vs. Predicted Rank Comparison

Retrieve the actual market cap at the target date (30th trading day after March 31, 2025). Assign descending ranks to both the predicted and actual market caps, then compute the absolute rank error for each stock.

```python
# 30th trading day after 2025-03-31 ≈ April 30, 2025
target_idx = min(STEPS - 1, len(test) - 1)
actual_at_target = test.iloc[target_idx]
print(f"Target date used: {test.index[target_idx].date()}")

pred_s   = pd.Series(predictions)
actual_s = actual_at_target.reindex(pred_s.index)

comparison = pd.DataFrame({
    "predicted": pred_s,
    "actual":    actual_s
}).dropna()

comparison["pred_rank"]   = comparison["predicted"].rank(ascending=False)
comparison["actual_rank"] = comparison["actual"].rank(ascending=False)
comparison["rank_error"]  = (comparison["pred_rank"] - comparison["actual_rank"]).abs()

print(f"Stocks in final comparison: {len(comparison)}")
comparison.head(20)
```

### Step 4 — Rank Error Metrics

Summarise prediction quality with two statistics:

- **Mean Absolute Rank Error:** average displacement in rank position between predicted and actual
- **Spearman ρ:** rank correlation between predicted and actual prices across all stocks (1 = perfect, 0 = no signal)

A histogram shows the full distribution of rank errors.

```python
mae_rank = comparison["rank_error"].mean()
rho, pval = spearmanr(comparison["predicted"], comparison["actual"])

print(f"Stocks evaluated      : {len(comparison)}")
print(f"Mean Abs Rank Error   : {mae_rank:.1f}")
print(f"Spearman ρ            : {rho:.4f}  (p={pval:.4f})")

comparison["rank_error"].hist(bins=50)
plt.xlabel("Absolute Rank Error")
plt.ylabel("Count")
plt.title("Distribution of Rank Errors (AR10, 30-day forecast to ~Apr 30 2025)")
plt.tight_layout()
plt.show()
```

## Diagnostic Plots

### Plot 1 — Predicted vs Actual: Market Cap & Rank

Two scatter plots side-by-side:
- **Left:** predicted market cap vs actual market cap with a 45° reference line — any systematic over/under-prediction shows as deviation from the diagonal.
- **Right:** predicted rank vs actual rank — a visual representation of Spearman ρ. Perfect ranking would lie on the diagonal.

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# --- Market cap scatter ---
ax = axes[0]
ax.scatter(comparison["actual"], comparison["predicted"], alpha=0.25, s=8, color="steelblue")
lim_lo = min(comparison["actual"].min(), comparison["predicted"].min())
lim_hi = max(comparison["actual"].quantile(0.99), comparison["predicted"].quantile(0.99))
ax.plot([lim_lo, lim_hi], [lim_lo, lim_hi], "r--", lw=1, label="y = x")
ax.set_xlim(lim_lo, lim_hi)
ax.set_ylim(lim_lo, lim_hi)
ax.set_xlabel("Actual Market Cap (~Apr 30, 2025)")
ax.set_ylabel("Predicted Market Cap")
ax.set_title("Predicted vs Actual Market Cap")
ax.legend()

# --- Rank scatter ---
ax = axes[1]
ax.scatter(comparison["actual_rank"], comparison["pred_rank"], alpha=0.25, s=8, color="darkorange")
n = len(comparison)
ax.plot([1, n], [1, n], "r--", lw=1, label="perfect ranking")
ax.set_xlabel("Actual Rank")
ax.set_ylabel("Predicted Rank")
ax.set_title(f"Predicted vs Actual Rank  (Spearman ρ = {rho:.3f})")
ax.legend()

plt.tight_layout()
plt.show()
```

### Plot 3 — Rank Error by Size Decile & CDF

- **Left bar chart:** mean absolute rank error split into 10 deciles by actual rank (decile 1 = largest-cap stocks). A U-shaped pattern would indicate the model struggles most at the extremes.
- **Right CDF:** cumulative % of stocks within a given rank error. Vertical dashed lines at ±50, ±100, ±250 show what fraction of stocks are "close enough" for practical rebalancing purposes.

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# --- Rank error by size decile ---
ax = axes[0]
comparison["size_decile"] = pd.qcut(comparison["actual_rank"], q=10, labels=range(1, 11))
decile_mae = comparison.groupby("size_decile", observed=True)["rank_error"].mean()
decile_mae.plot(kind="bar", ax=ax, color="steelblue", edgecolor="white")
ax.set_xlabel("Actual Rank Decile  (1 = largest cap)")
ax.set_ylabel("Mean Absolute Rank Error")
ax.set_title("Rank Error by Size Decile")
ax.tick_params(axis="x", rotation=0)

# --- CDF of rank errors ---
ax = axes[1]
sorted_err = np.sort(comparison["rank_error"].values)
cdf = np.arange(1, len(sorted_err) + 1) / len(sorted_err) * 100
ax.plot(sorted_err, cdf, color="darkorange", lw=2)
for thresh in [50, 100, 250]:
    pct = (comparison["rank_error"] <= thresh).mean() * 100
    ax.axvline(thresh, color="gray", linestyle="--", lw=0.8)
    ax.text(thresh + 8, 5, f"{pct:.0f}% ≤ {thresh}", fontsize=8, color="gray")
ax.set_xlabel("Absolute Rank Error")
ax.set_ylabel("Cumulative % of Stocks")
ax.set_title("CDF of Rank Errors")
ax.set_ylim(0, 100)

plt.tight_layout()
plt.show()
```

### Plot 4 — Russell Index Membership Overlap

For each Russell index, what percentage of actual members did we correctly predict?

| Index | Membership band |
|---|---|
| Russell 1000 | Top 1000 by market cap (large cap) |
| Russell 2000 | Ranks 1001–3000 (small cap) |
| Russell 3000 | Top 3000 (broad market) |
| Russell Microcap | Ranks 2001–4000 (micro cap) |

```python
RUSSELL_BANDS = {
    "Russell 1000":     (1,    1000),
    "Russell 2000":     (1001, 3000),
    "Russell 3000":     (1,    3000),
    "Russell Microcap": (2001, 4000),
}

def in_band(rank_col, lo, hi):
    return (rank_col >= lo) & (rank_col <= hi)

overlaps, labels = [], []
for name, (lo, hi) in RUSSELL_BANDS.items():
    pred_in   = set(comparison[in_band(comparison["pred_rank"],   lo, hi)].index)
    actual_in = set(comparison[in_band(comparison["actual_rank"], lo, hi)].index)
    n_actual  = len(actual_in)
    correct   = len(pred_in & actual_in)
    pct       = correct / n_actual * 100 if n_actual > 0 else 0
    overlaps.append(pct)
    labels.append(name)
    print(f"{name:20s}: {pct:.1f}%  ({correct}/{n_actual} correctly identified)")

fig, ax = plt.subplots(figsize=(9, 4))
ax.bar(labels, overlaps, color="steelblue", edgecolor="white")
for i, pct in enumerate(overlaps):
    ax.text(i, pct + 1, f"{pct:.1f}%", ha="center", fontsize=10)
ax.axhline(100, color="green", linestyle="--", lw=1, label="Perfect (100%)")
ax.set_ylabel("Membership Overlap (%)")
ax.set_ylim(0, 110)
ax.set_title("Russell Index Membership: Predicted vs Actual\n(% of actual members correctly predicted)")
ax.legend()
plt.tight_layout()
plt.show()
```

## Rebalance Add/Remove Prediction

Russell reconstitutes annually in late June. The most recent reconstitution before our April 30, 2025 target was **June 2024**. Using market cap ranks as of June 28, 2024 as a proxy for prior index membership, we can ask:

> Of the stocks that *actually* changed membership status by April 30, 2025 — how many did our AR(10) model correctly flag?

For each index we compute:
- **Adds**: stocks outside the band in June 2024 that entered by April 30, 2025
- **Removes**: stocks inside the band in June 2024 that exited by April 30, 2025
- **Precision**: of our predicted adds/removes, what fraction were real?
- **Recall**: of the real adds/removes, what fraction did we catch?

```python
# Proxy for June 2024 reconstitution — nearest available date in training data
prev_rebal = pd.Timestamp("2024-06-28")
prev_date  = train.index[train.index.get_indexer([prev_rebal], method="nearest")[0]]
print(f"Previous rebalance proxy date: {prev_date.date()}")

prev_mktcap = train.loc[prev_date].reindex(comparison.index)
comparison["prev_rank"] = prev_mktcap.rank(ascending=False)

rows = []
for name, (lo, hi) in RUSSELL_BANDS.items():
    prev_in   = in_band(comparison["prev_rank"],   lo, hi)
    pred_in   = in_band(comparison["pred_rank"],   lo, hi)
    actual_in = in_band(comparison["actual_rank"], lo, hi)

    actual_adds    = (~prev_in) & actual_in
    actual_removes = prev_in   & (~actual_in)
    pred_adds      = (~prev_in) & pred_in
    pred_removes   = prev_in   & (~pred_in)

    tp_adds    = (pred_adds    & actual_adds).sum()
    tp_removes = (pred_removes & actual_removes).sum()
    n_aa, n_ar = actual_adds.sum(), actual_removes.sum()
    n_pa, n_pr = pred_adds.sum(),   pred_removes.sum()

    rows.append({
        "Index":            name,
        "Actual Adds":      n_aa,
        "Pred Adds":        n_pa,
        "Add TP":           tp_adds,
        "Add Precision":    tp_adds / n_pa if n_pa > 0 else 0,
        "Add Recall":       tp_adds / n_aa if n_aa > 0 else 0,
        "Actual Removes":   n_ar,
        "Pred Removes":     n_pr,
        "Remove TP":        tp_removes,
        "Remove Precision": tp_removes / n_pr if n_pr > 0 else 0,
        "Remove Recall":    tp_removes / n_ar if n_ar > 0 else 0,
    })

rebal_df = pd.DataFrame(rows).set_index("Index")

fmt = {c: "{:.1%}" for c in rebal_df.columns if "Precision" in c or "Recall" in c}
rebal_df.style.format(fmt)
```

### Visualisation — Adds & Removes by Index

Each subplot shows actual vs predicted add/remove counts for one Russell index. The precision and recall annotation in the top-right corner summarises classification quality. A perfect model would have bars of equal height and 100% precision and recall.

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

for ax, (name, _) in zip(axes.flatten(), RUSSELL_BANDS.items()):
    row = rebal_df.loc[name]
    w = 0.3
    cats = ["Adds", "Removes"]
    actual_counts = [row["Actual Adds"],  row["Actual Removes"]]
    pred_counts   = [row["Pred Adds"],    row["Pred Removes"]]
    tp_counts     = [row["Add TP"],       row["Remove TP"]]

    x = [0, 1]
    ax.bar([i - w/2 for i in x], actual_counts, w, label="Actual",    color="steelblue",  alpha=0.85)
    ax.bar([i + w/2 for i in x], pred_counts,   w, label="Predicted", color="darkorange", alpha=0.85)
    for i, (tp, act) in enumerate(zip(tp_counts, actual_counts)):
        if act > 0:
            ax.text(i - w/2, act + 0.5, f"TP={tp}", ha="center", fontsize=8, color="steelblue")

    ax.set_xticks(x)
    ax.set_xticklabels(cats)
    ax.set_title(name)
    ax.set_ylabel("Number of Stocks")
    ax.legend(fontsize=8)

    ann = (f"Adds:    P={row['Add Precision']:.0%}  R={row['Add Recall']:.0%}\n"
           f"Removes: P={row['Remove Precision']:.0%}  R={row['Remove Recall']:.0%}")
    ax.text(0.98, 0.97, ann, transform=ax.transAxes, fontsize=8, va="top", ha="right",
            family="monospace", bbox=dict(boxstyle="round", facecolor="white", alpha=0.8))

plt.suptitle("Predicted vs Actual Membership Changes\n(relative to June 2024 rebalance)", y=1.01, fontsize=13)
plt.tight_layout()
plt.show()
```
